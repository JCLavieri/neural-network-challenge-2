```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
data = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
data.head()
```

```python
# Determine the number of unique values in each column.
data.nunique()
```

```python
print(data.columns)
```

```python
# Create y_df with the Attrition and Department columns
y_df = data[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
X_df = data





# Create X_df using your selected columns
X_df.head()


# Show the data types for X_df
X_df.dtypes
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=42)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Define preprocessing for numeric and categorical data
numeric_features = X_df.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X_df.select_dtypes(include=['object']).columns

# Create transformers
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Combine transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

# Apply transformations
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
# OneHot encode 'department'
encoder_dept = OneHotEncoder()
y_train_dept = encoder_dept.fit_transform(y_train[['Department']])
y_test_dept = encoder_dept.transform(y_test[['Department']])

# OneHot encode 'attrition'
encoder_attr = OneHotEncoder()
y_train_attr = encoder_attr.fit_transform(y_train[['Attrition']])
y_test_attr = encoder_attr.transform(y_test[['Attrition']])
```

Part 2: Create, Compile, and Train the Model

1. Find Number of Columns in X Training Data

```python
n_features = X_train.shape[1]
n_features
```

2. Create Input Layer and Shared Layers

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# Input layer
input_layer = Input(shape=(n_features,))

# Shared layers
shared1 = Dense(128, activation='relu')(input_layer)
shared2 = Dense(64, activation='relu')(shared1)
```

3. & 4. Create Output Branches

```python
# Department branch
dept_branch = Dense(32, activation='relu')(shared2)
dept_output = Dense(y_train_dept.shape[1], activation='softmax', name='department_output')(dept_branch)

# Attrition branch
attr_branch = Dense(32, activation='relu')(shared2)
attr_output = Dense(y_train_attr.shape[1], activation='sigmoid', name='attrition_output')(attr_branch)
```

5. Create and Compile the Model

```python
model = Model(inputs=input_layer, outputs=[dept_output, attr_output])
model.compile(optimizer='adam', loss=['categorical_crossentropy', 'binary_crossentropy'], metrics=['accuracy'])
```

6. Summarize the Model

```python
model.summary()
```

7. Train the Model

```python
history = model.fit(X_train, [y_train_dept.toarray(), y_train_attr.toarray()], validation_split=0.2, epochs=100, batch_size=32)
```

8. Evaluate the Model

```python
loss, dept_loss, attr_loss, dept_acc, attr_acc = model.evaluate(X_test, [y_test_dept.toarray(), y_test_attr.toarray()])
print(f"Department Accuracy: {dept_acc}, Attrition Accuracy: {attr_acc}")
```
